```python
import pandas as pd
```

```python
df = pd.read_csv("musicnet_metadata_periods(Completed).csv", encoding="ISO-8859-1")

df.head()
```

```python
for index, row in df.iterrows():
    id=str(row["unique id"])
    period=id.split(".")[0]
    path="Musicnet+/"+period+"/"+id+".wav"
    print(path)
    df.loc[index, 'path']=path
```
